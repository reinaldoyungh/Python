```python
import google.generativeai as genai
from google.colab import userdata
import time # Importando o módulo time para usar time.sleep()


# 1. Configuração da API (Recuperando a chave do gerenciador de segredos do Colab)
genai.configure(api_key=userdata.get('GEMINI_API_KEY'))
model = genai.GenerativeModel('gemini-2.5-flash')

def resumir_emails_com_gemini(lista_emails):
    resumos = []

    print(f"Iniciando o processamento de {len(lista_emails)} e-mails...\n")

    for i, corpo in enumerate(lista_emails, 1):
        # Criando o prompt específico para a tarefa
        prompt = f"Resuma o desejo/intenção do remetente deste e-mail em apenas uma linha curta:\n\n{corpo}"

        try:
            # Chamada para a API
            response = model.generate_content(prompt)
            resumo_linha = response.text.strip()

            print(f"E-mail {i} processado com sucesso.")
            resumos.append(resumo_linha)

        except Exception as e:
            print(f"Erro ao processar o e-mail {i}: {e}")
            resumos.append("Erro ao gerar resumo.")

        # Adiciona um atraso de 60 segundos entre as requisições
        time.sleep(20)

    return resumos

# --- Exemplo de Uso ---

# Supondo que a variável 'corpos_email' (aquela lista de 20) já exista no seu código:
lista_de_resumos = resumir_emails_com_gemini(corpos_email)

print("\n--- RESULTADOS ---")
for numero, r in enumerate(lista_de_resumos):
    print(f"Email {numero + 1}: {r}")
    print("-" * 50)
```
