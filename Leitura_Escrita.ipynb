# Lendo e escrevendo arquivos com Python

```python
corpos_email = [
    "Olá equipe,\n\nGostaria de agendar nossa reunião de alinhamento para discutir os resultados do último trimestre e planejar os próximos passos. Por favor, confirmem a disponibilidade na agenda compartilhada.",

    "Fala, pessoal!\n\nPara comemorar as metas batidas, vamos nos reunir no bar da esquina nesta sexta às 18h. Quem topa?\n\nAté lá!",

    "Prezados colaboradores,\n\nInformamos que houve uma atualização nas diretrizes de trabalho remoto. O documento em anexo detalha as novas regras que entram em vigor a partir do próximo mês."]
```

```python
import google.generativeai as genai
from google.colab import userdata


# 1. Configuração da API (Recuperando a chave do gerenciador de segredos do Colab)
genai.configure(api_key=userdata.get('GEMINI_API_KEY'))
model = genai.GenerativeModel('gemini-2.5-flash')

def resumidor_de_emails(lista_de_emails):
    lista_de_resumos = []

    for numero, email in enumerate(lista_de_emails):
        resposta = model.generate_content(
            contents=f"""Vou te mandar o corpo de um e-mail. Quero que você o resuma em apenas 1 linha,
            passando o intuito daquele e-mail. Segue o e-mail: {email}"""
        )
        print(f"E-mail {numero + 1}: {resposta.text}")
        lista_de_resumos.append(f"E-mail {numero + 1}: {resposta.text}")
        lista_de_resumos.append("-" * 50)
        print("-" * 50)

    return lista_de_resumos
```

```python
lista_resumos = resumidor_de_emails(corpos_email)
```

```python
lista_resumos
```

```python
with open('lista-de-resumos.txt', 'w',encoding="utf-8") as arquivo:
  for resumo in lista_resumos:
    arquivo.write(resumo + '\n')
```

# Leitura de arquivos com o Python

```python
nova_lista_leitura = []
with open('lista-de-resumos.txt', 'r',encoding="utf-8") as arquivo:
  for linha in arquivo:
    nova_lista_leitura.append(linha.strip())

nova_lista_leitura
```
